```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
x = df[['Monthly Grocery Bill',
        'Vehicle Monthly Distance Km',
        'How Long TV PC Daily Hour',
        'How Many New Clothes Monthly',
        'How Long Internet Daily Hour']]
y = df['CarbonEmission']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(max_depth=10, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, max_depth=5, random_state=42),
    'K-Nearest Neighbors': KNeighborsRegressor(n_neighbors=5),
    'Support Vector Machine': SVR(kernel='rbf', C=100, gamma=0.1)
}
```

```python
results = []
best_model = None
best_r2 = -float('inf')

print("=" * 80)
print("CARBON FOOTPRINT PREDICTION - MODEL COMPARISON")
print("=" * 80)
    
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='r2')
    cv_mean = cv_scores.mean()
    
    results.append({
        'Model': name,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R2 Score': r2,
        'CV R2 Mean': cv_mean
    })
    
    print(f"\n{name}:")
    print(f"  MSE:      {mse:,.2f}")
    print(f"  RMSE:     {rmse:,.2f}")
    print(f"  MAE:      {mae:,.2f}")
    print(f"  R² Score: {r2:.4f}")
    print(f"  CV R² Mean: {cv_mean:.4f}")
    
    if r2 > best_r2:
        best_r2 = r2
        best_model = model
        best_model_name = name
```

```python
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('R2 Score', ascending=False)

print("\n" + "=" * 80)
print("RANKING BY R² SCORE")
print("=" * 80)
print(results_df.to_string(index=False))
```

```python
with open('best_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)

results_df.to_csv('model_comparison.csv', index=False)

with open('best_model_info.txt', 'w') as f:
    f.write(f'Best Model: {best_model_name}\n')
    f.write(f'R² Score: {best_r2:.4f}\n')
    f.write(f'RMSE: {np.sqrt(mean_squared_error(y_test, best_model.predict(x_test))):,.2f}\n')
    f.write(f'MAE: {mean_absolute_error(y_test, best_model.predict(x_test)):,.2f}\n')
    f.write(f'\nModel trained on {len(x_train)} samples\n')
    f.write(f'Tested on {len(x_test)} samples\n')

print(f"\n{'=' * 80}")
print(f"BEST MODEL: {best_model_name} (R² = {best_r2:.4f})")
print(f"Model saved as 'best_model.pkl'")
print(f"{'=' * 80}")
```
